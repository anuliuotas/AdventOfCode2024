```python
lines = []
with open('input.txt', 'r') as file:
    for line in file:
        lines.append(line.replace('\n', '')) 
```

```python
lines[:5]
```

```python
max_col = len(lines[0])
max_row = len(lines)
cols = [[] for _ in range(max_col)]
rows = [[] for _ in range(max_row)]
fdiag = [[] for _ in range(max_row + max_col - 1)]
bdiag = [[] for _ in range(len(fdiag))]
min_bdiag = -max_row + 1

for x in range(max_col):
    for y in range(max_row):
        cols[x].append(lines[y][x])
        rows[y].append(lines[y][x])
        fdiag[x+y].append(lines[y][x])
        bdiag[x-y-min_bdiag].append(lines[y][x])

print(cols)
print(rows)
print(fdiag)
print(bdiag)
```

```python
to_check = []
for ch in [cols, rows, fdiag, bdiag]:
    for line in ch:
        line_str = ''.join(line)
        to_check.append(line_str)
        to_check.append(line_str[::-1])
```

```python
count = 0

for line in to_check:
    count += len(line.split('XMAS')) - 1
count
```
